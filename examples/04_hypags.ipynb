```python
import pedon as pe

import matplotlib.pyplot as plt
```

```python
ss = pe.SoilSample(
    k=1e-5  # m/s
)
hyp = ss.hypags()
```

```python
f, axs = plt.subplots(1, 2, figsize=(8, 6), sharey=True, layout="tight")

pe.plot_swrc(hyp, ax=axs[0], label="HYPAGS")
axs[0].set(
    yscale="log",
    title="Soil Water Retention Curve",
    xlabel="\N{GREEK SMALL LETTER THETA} [-]",
    xlim=(0, 0.3),
    ylabel="|\N{GREEK SMALL LETTER PSI}| [cm]",
)
axs[0].legend(loc="lower left")

pe.plot_hcf(hyp, ax=axs[1], label="HYPAGS")
axs[1].set(
    yscale="log",
    xscale="log",
    title="Hydraulic Conductivity Function",
    xlabel="Ks [m/s]",
    xlim=(1e-12, 1e-4),
)
axs[1].legend(loc="lower left")
```
